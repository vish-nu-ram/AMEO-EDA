Exploratory data analytics on AMEO data set

Aspiring Minds’ Employability Outcomes 2015 (AMEO 2015) is a unique dataset which contains engineering graduates’ employment outcomes (salaries, job titles and job locations) along with standardized assessment scores in three fundamental areas - cognitive skills, technical skills and personality. 

Data source: http://research.aspiringminds.com/resources/#ameo

# Corelation between pair of values


```
sns.jointplot('Salary', 'collegeGPA',hue='Gender' ,data=df.sample(500))
sns.jointplot('Salary', '12percentage', hue='Gender' ,data=df.sample(500))
sns.jointplot('Salary', 'GraduationYear', hue='Gender' ,data=df.sample(500))
```

There is no corelation for salary with variables such as College GPA, 12th percentage or year of graduation. most of the salary is saturated below 6 lakhs irrespective of GPA

To get a cleared picture of the bulk of the data, let us focus on only salaries less than 500000. 
Again here we see no corelation whatsoever.

```
df_ = df[(df.Salary<500000)]
sns.jointplot('Salary', 'collegeGPA', data=df_.sample(500))
sns.jointplot('Salary', '12percentage', data=df_.sample(500))
sns.jointplot('Salary', 'GraduationYear', data=df_.sample(500))
```

#Gender Gap

```
 sns.violinplot(x='Gender',y='Salary', data = df.sample(500))
```

Here we can observe that in the medium range, Both males and females earn similar salaries, Seems like more males are paid higher salaries.

Let us fetch Statistics values to compare

```
m = df[(df.Gender =='m')]
f = df[(df.Gender == 'f')]

mean = (np.mean(m.Salary),np.mean(f.Salary))
median = (np.median(m.Salary),np.median(f.Salary))
```

```
sns.barplot(x = 'Gender', y = df.Salary, estimator=np.mean, data = df)
```

Mean values for Males is slightly higher

```
df_ = df[(df['Salary']>700000)]
sns.barplot(x = 'Gender', y = df.Salary, estimator=np.mean, data = df_)
```

Among the higher Salaries, Females are paid slightly higher on an average.
